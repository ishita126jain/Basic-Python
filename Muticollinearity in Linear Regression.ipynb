```python
import pandas as pd
```

```python
import statsmodels.api as sm
```

```python
df_adv = pd.read_csv('/home/ishita/Documents/datasets/Advertising.csv')
```

```python
df_adv
```

```python
x = df_adv[['TV','Radio','Newspaper']]
y = df_adv['Sales']
```

```python
# fit a OLS (ordinary least square) 
# complete equation of linear regression is β0 + β1(TV) +β2(Radio) +β3(Newspaper)
# For OLS model we need to obtain β0 also we multiply β0 by 1.
# Then equation become β0(1) + β1(TV) +β2(Radio) +β3(Newspaper).
# For this we need to one column with value 1.
# Lets create a column of constant 1 by using add_constant function of statsmodels.
x = sm.add_constant(x)
```

```python
x
```

```python
# In sm.OLS(need to give output feature , need to give input feature).
model = sm.OLS(y,x).fit()
```

```python
model.summary()
```

```python
import matplotlib.pyplot as plt
```

```python
x.iloc[:,1:].corr()
```

```python
# we can see that in above table there is no collinearity between the independent features
# so there is no multicollinearity issue in this dataset.
```

```python
# Now we check multicollinearity in another dataset
df_salary = pd.read_csv('/home/ishita/Documents/datasets/Salary_Data.csv')
```

```python
df_salary
```

```python
x = df_salary[['YearsExperience','Age']]
y = df_salary['Salary']
```

```python
x = sm.add_constant(x)
```

```python
x
```

```python
model = sm.OLS(y,x).fit()
```

```python
model.summary()
```

```python
# First a fall we see a coeff here in this case the value of coeff is very large number meanse there is a very large change on changing them.
# Second we check the R squared ans adj R squared value it is 96 95 percent meanse our modeltrained good enough.
# Then we see std err and here std err is very large meanse there is a change of collinearity.
# At last we focus on p value which sould less than 0.05 here p value of experience is less than 0.05 but age have p value more than 0.1 which is too much so ther is a chance of multicolinearity.
# Lets check it using corr()
```
