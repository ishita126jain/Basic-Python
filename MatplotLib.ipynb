```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
plt.scatter(x,y,c='g')

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('2D graph')
```

```python
y=x*x
```

```python
plt.plot(x,y,'ro--')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'y*--')
plt.subplot(2,2,3)
plt.plot(x,y,'go')
```

```python
np.pi
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)

plt.title("sin wave")
plt.plot(x,y)
```

```python
x=np.arange(0,4*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

plt.subplot(2,1,1)
plt.title("sin wave")
plt.plot(x,y_sin,'r')

plt.subplot(2,1,2)
plt.title("cos wave")
plt.plot(x,y_cos,'g')
```

```python
x=[2,8,10]
y=[11,16,9]

a=[3,9,11]
b=[6,15,7]

plt.bar(x,y)
plt.bar(a,b,color='g')

plt.title("Bar graph")
plt.ylabel('Y axis')
plt.xlabel('X axis')
```

```python
a=np.array([22,87,5,43,56,73,55,11,20,51,5,79,31,27])
plt.hist(a,color='y')
plt.title('histogram')
```

```python
data = [np.random.normal(0,std,100)for std in range(1,4)]

plt.boxplot(data,vert=True,patch_artist=True)
```

```python
labels = 'Python','C++','Ruby','Java'
sizes = [215,130,245,210]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.1,0,0,0)

plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)
plt.axis('equal')
```
